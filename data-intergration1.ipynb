```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# 주어진 칼럼들의 데이터
columns = ['소멸지수', '병상 비율', '병원', '보건지소', '약국', '요양병원', '인구 수', '종합병원', '고등학교', 
           '보육시설 비율', '유치원', '중학교', '초등학교', '전규모W', '1인당 자동차 등록대수', '소규모C', 
           '중규모C', '대규모C', '소규모W', '중규모W', '대규모W']

# Q-Q 플롯 그리기
for col in columns:
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111)
    sm.qqplot(np.array(df[col]), line='s', ax=ax)
    ax.set_title(f'Q-Q Plot for {col}')
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# 시군구를 제외한 칼럼 선택
columns_to_scale = ['소멸지수', '병상 비율', '병원', '보건지소', '약국', '요양병원', '인구 수', '종합병원', 
                   '보육시설 비율', '유치원', '중학교', '초등학교', '전규모W', '1인당 자동차 등록대수', 
                   '소규모C', '중규모C', '대규모C', '소규모W', '중규모W', '대규모W']

# 이상치 제거
for column in columns_to_scale:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]

# Min-Max 스케일링
scaler = MinMaxScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# 스케일링 결과 확인
df
```

```python
import pandas as pd
from scipy.stats import shapiro
import numpy as np
shapiro_df = df[['소멸지수', '병상 비율', '병원', '보건지소', '약국', '요양병원', '인구 수', '종합병원', 
                   '보육시설 비율', '유치원', '중학교', '초등학교', '전규모W', '1인당 자동차 등록대수', 
                   '소규모C', '중규모C', '대규모C', '소규모W', '중규모W', '대규모W']]

# 샤피로-윌크 검정 결과를 저장할 리스트 초기화
results = []

# 각 열에 대해 샤피로-윌크 검정 수행 및 결과 저장
for column in shapiro_df.columns:  # 각 열에 대해
    stat, p_value = shapiro(shapiro_df[column])
    
    # 결과 저장
    results.append({
        'Column': column,
        'Statistic': stat,
        'p-value': p_value,
        '정규 분포 여부': '정규 분포' if p_value > 0.05 else '비정규 분포'
    })

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results)

# 결과 출력
results_df
```
